```python
# import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# read the datatset into pandas
df = pd.read_csv('marketing_data.csv')
```

```python
df.head()
```

```python
df.tail(10)
```

```python
df.shape
```

```python
df.info()
```

#### There is only one column with null values in the dataset

```python
df.isna().sum()

#the income column has 24 null values
```

```python
# replacing the null values with zero
df['Income'].fillna(0, inplace = True)
```

```python
df.isna().sum()
```

#### There are now zero null values in all of the dataset 

```python
# making the columns into a list
df.columns.tolist()
```

```python
# number of unique values in each column
df.nunique()
```

```python
# creating a new column for the Age of the customers
def cust_age(year):
    age = 2023 - year
    return age
Ages = cust_age(df['Year_Birth'])
Ages
```

```python
# making age a new column in our dataset
df['Cust_age'] = Ages
```

```python
df.head()
```

```python
# checking for the unique values in the age column
df['Cust_age'].unique()
```

```python
# checking for the maximum and minimum(range) age
age_range = [df['Cust_age'].min(), df['Cust_age'].max()]
age_range
```

```python
# grouping the ages into bands
def age_band(age):
    if age >= 25 and age <= 35:
        ageband = '25-35'
    elif age > 35 and age <= 50:
        ageband = '35-50'
    else:
        ageband = '>50'
    return ageband

df['Age_band'] = df['Cust_age'].apply(age_band)
```

```python
df.head()
```

```python
# combining Kidhome and Teenhome into one column
df['Kids'] = df[['Kidhome','Teenhome']].sum(axis = 1)
df.drop(['Kidhome','Teenhome'], axis = 1, inplace = True)
```

```python
# combining all goods as one
df['Tot_amt_goods'] = df[['MntWines','MntFruits','MntMeatProducts','MntFishProducts',
                          'MntSweetProducts','MntGoldProds']].sum(axis = 1)
```

```python
df.head()
```

```python
# the distribution of marital status according to their volume of purchase
vol_goods = df.groupby('Marital_Status').sum()['Tot_amt_goods']
vol_goods
```

```python
# visualize the distribution of marital status and purchase volume
vol_goods.plot(kind = 'line', marker = 'o')
plt.title('Volume of Goods Purchased per Marital Status')
plt.xlabel('marriage_status')
plt.ylabel('volume_goods')
plt.xticks(rotation = 45)

plt.show()
```

```python
# the average distribution of income according to their marital status
av_distro = df.groupby('Marital_Status').mean()['Income'].sort_values(ascending = False)
av_distro
```

```python
# visualize the average distribution of income according to their marital status
av_distro.plot(kind='bar', color='green')
plt.title('Average Distribution of Income per Marital Status')
plt.xticks(rotation=45)

plt.show()
```

```python
# the distribution of goods purchased according to their age band
top_spenders = df.groupby('Age_band').sum()['Tot_amt_goods']
top_spenders
```

```python
# visualize the distribution of goods purchased according to their age band
P_explode = [0, 0.2, 0.2]
color = ["#3A5F0B", '#228B22', "#50C878"]
# "#3A5F0B", "#50C878", '#1AA260','#228B22'
top_spenders.plot(kind = 'pie', explode = P_explode, colors = color, shadow = True, autopct='%1.1f%%')
```

```python
df.head()
```

```python
# the distribution of age band according to their value of income
age_per_income = df.groupby('Age_band').mean()['Income']
age_per_income
```

```python
# the average income of each band according to their countries
average_income = df.pivot_table(index = 'Age_band', columns = 'Country', values = 'Income')
average_income
```

```python
# visualize the average income of each band according to their countries
average_income.plot(kind = 'bar', stacked=True, color = ['#1B8A6B', '#1AA260', '#50C878', '#617C58',
                                                        '#728C00','#808000','#3A5F0B','#4B5320'])
plt.title('Average Income of Each Age Bands per Country')
plt.legend(loc='upper left', fontsize=9, bbox_to_anchor=(1, 0.5))
plt.xticks(rotation=360)

plt.show()
```

```python
# total amount of goods purchased by country according to the amount of kids they have
df.pivot_table(index = 'Kids', columns = 'Country', values = 'Tot_amt_goods', aggfunc = 'sum')
```

```python
# pivot tables gives you the average values by default except you specify otherwise
# the average amount of goods purchased by each age band according to their countries
avg_purchase = df.pivot_table(index = 'Age_band', columns = 'Country', values = 'Tot_amt_goods')
avg_purchase 
```

```python
# visualize average amount of goods purchased by each age band per country
avg_purchase.plot(kind = 'bar', stacked=True, color = ['#1B8A6B', '#1AA260', '#50C878', '#617C58',
                                                        '#728C00','#808000','#3A5F0B','#4B5320'])
plt.title('Average Amount of Goods Purchased by Each Age Band per Country')
plt.legend(loc='upper left', fontsize=9, bbox_to_anchor=(1, 0.5))
plt.xticks(rotation=360)

plt.show()
```

```python
df['Education'].unique()
```

```python
# average income of each age band by their Education status
avg_income = df.pivot_table(index = 'Age_band', columns = 'Education', values = 'Income')
avg_income
```

```python
# visualize average income of each age band by Education status
avg_income.plot(kind = 'bar', stacked=True, color = ['#1B8A6B', '#1AA260', '#50C878', '#617C58','#728C00'])
plt.title('Average Income of Each Age Band according per Education Status')
plt.legend(loc='upper left', fontsize=9, bbox_to_anchor=(1, 0.5))
plt.xticks(rotation=360)

plt.show()
```

```python
# total goods purchased by each age band according to their Education status
total_goods = df.pivot_table(index = 'Age_band', columns = 'Education', values = 'Tot_amt_goods', aggfunc = 'sum')
total_goods
```

```python
# visualize the goods purchased by each age band according by Education status
total_goods.plot(kind = 'bar', stacked=True, color = ['#1B8A6B', '#1AA260', '#50C878', '#617C58','#728C00'])
plt.title('Total Goods Purchased by Age Band per Education Status')
plt.legend(loc='upper left', fontsize=9, bbox_to_anchor=(1, 0.5))
plt.xticks(rotation=360)

plt.show()
```

## INSIGHTS

#### From the data we derived the following insights:

- Married people purchased the most amount of goods while YOLO category purchased the least amount.
- In total, people above 50 years old purchase more goods than the other age bands.
- In Germany, the age band from 25 to 35 years earn the highest income.
- In Saudi Arabia, the age band from 35 to 50 years earn the highest income.
- In Saudi Arabia and Spain, the age band from 50 and above earn the highest income.
- In Mexico, customers purchased the highest average amount of goods in all the age bands.
- Generally, the Graduation Education Status purchased more goods for all Age bands.
